```python


fig=plt.figure(figsize=(4.5,3))    
Oscillator3 = PhasePortrait2D(ground_truth, Range=[[-4,4], [-2, 2]], MeshDim=21,Title="", xlabel=r"$x_1$", ylabel=r"$x_2$", fig=fig)


fig, ax=Oscillator3.plot(color='Greys')


mu_x, mu_y = 3, 0  # Mean of the Gaussian distribution
sigma_x, sigma_y = 0.8, 0.8  # Standard deviation of the Gaussian distribution
size = 70  # Size of the grid

# Create a grid of x and y values within the specified range
x = np.linspace(-4, 4, size)
y = np.linspace(-1.5, 1.5, size)
X, Y = np.meshgrid(x, y)
threshold=0.05
# Plot the discretized Gaussian heatmap
# Compute the Gaussian distribution at each point on the grid
Z = (1 / (2 * np.pi * sigma_x * sigma_y)) * np.exp(-((X - mu_x)**2 / (2 * sigma_x**2) + (Y - mu_y)**2 / (2 * sigma_y**2)))
x_pts, y_pts = np.where(Z > threshold)
ax.scatter(X[x_pts, y_pts], Y[x_pts, y_pts], c=Z[x_pts, y_pts], cmap='Blues', s=10, marker='o')
# fig.set_size_inches(5, 3.5)
fig.suptitle("Uncertainty distribution over data", fontsize=16)
# plt.show()
fname = os.path.join(eq_idx+"_phase.density.pdf")
plt.savefig(fname, bbox_inches='tight', pad_inches=0)
plt.close()
```
